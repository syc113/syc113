# Loading the datase

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

# Model

```python
k_range=range(1,31)
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
x=iris.data
y=iris.target
data=[]
for k in k_range:  
    knn=KNeighborsClassifier(n_neighbors=k)  
    scores=cross_val_score(knn, x, y, cv=10)
    data.append([scores.mean()])
```

# Drawing

```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.DataFrame(data,columns=['scores'],dtype=float) 
plt.plot(df)
plt.grid()
plt.xlabel("Value of K for KNN")
plt.ylabel("Cross_Value scores")
plt.show() 
```

# Find optimal

```python
from sklearn.model_selection import GridSearchCV
import numpy as np
gs={'n_neighbors':np.arange(1,32)}
knn_gs=KNeighborsClassifier()
knn_param_search=GridSearchCV(knn_gs,gs,cv=10)
knn_param_search.fit(x,y)
knn_param_search.best_params_
```
